# 1. Growth Of Function
Time Complexity/Order of Growth defines the amount of time taken by any program with respect to the size of the input. Time Complexity specifies how the program would behave as the order of size of input is increased. So, Time Complexity is just a function of size of its input. We will go through some of basic and most common time complexities such as:

- *Constant Time Complexity O(1) : constant running time*
- *Linear Time Complexity O(n) : linear running time*
- *Logarithmic Time Complexity O(log n) : logarithmic running time*
- *Log-Linear Time Complexity O(n log n) : log-Â­linear running time*
- *Polynomial Time Complexity O(n^c) : polynomial running time (c is a constant)*
- *Exponential Time Complexity O(c^n) : exponential running time (c is a constant being raised to a power based on size of input)*

## Constant time Complexity:O(1)
An algorithm is said to have constant time with order O (1) when it is not dependent on the input size n. Irrespective of the input size n, the runtime will always be the same. See example below

```python
# Load the Python libraries
import matplotlib.pyplot as plt
```

```python
# Prob_1

def add_two_numbers(a, b):
    return a + b

n_values = list(range(1, 101))
time_values = []

for n in n_values:
    add_two_numbers(n, n + 1)
    time_values.append(1)

plt.figure()
plt.plot(n_values, time_values)
plt.xlabel("Input size n")
plt.ylabel("Time")
plt.title("O(1) Constant Time Complexity for Addition")
plt.show()
```

#Linear Time Complexity O(n) : linear running time
- *An algorithm is said to have a linear time complexity when the running time increases linearly with the length of the input. When the function involves checking all the values in input data, with this order O(n).*

```python
def linear_time_sum(numbers):
    total = 0
    for value in numbers:
        total += value
    return total

n_values = list(range(1, 101))
time_values = []

for n in n_values:
    data = list(range(n))
    linear_time_sum(data)
    time_values.append(n)

plt.figure()
plt.plot(n_values, time_values)
plt.xlabel("Input size n")
plt.ylabel("Time")
plt.title("O(n) Linear Time Complexity")
plt.show()
```

```python
import matplotlib.pyplot as plt

#Prob_3:Take another example and write the code that runs in linear amount of time no matter what is the size of input..
#similar to example-2, also plot the graph (eg. even_oddcounts elements greater than zero)

def count_greater_than_zero(numbers):
    count = 0
    for value in numbers:
        if value > 0:
            count += 1
    return count

n_values = list(range(1, 101))
time_values = []

for n in n_values:
    data = list(range(-n, n + 1)) # Create data with some numbers greater than zero
    count_greater_than_zero(data)
    time_values.append(n) # Simulate linear time

plt.figure()
plt.plot(n_values, time_values)
plt.xlabel("Input size n")
plt.ylabel("Time")
plt.title("O(n) Linear Time Complexity for Counting > 0")
plt.show()
```

#Logarithmic Time Complexity O(log n) : logarithmic running time
- *An algorithm is said to have a logarithmic time complexity when it reduces the size of the input data in each step. This indicates that the number of operations is not the same as the input size. The number of operations gets reduced as the input size increases. Algorithms are found in binary tree or binary search functions. This involves the search of a given value in an array by splitting the array into two and starting searching in one split. This ensures the operation is not done on every element of the data.*

```python
#Prob_4 Take Suitable example and write function that runs in logarithic time. plot graph as well similar to above example.
import math
import matplotlib.pyplot as plt

# Prob_4: Write a function that runs in logarithmic time (Binary Search)
def binary_search_time_simulation(n):

    if n == 0:  # Handle edge case for log(0) which is undefined
        return 0
    return math.log2(n)

n_values = list(range(1, 1001)) # Input sizes from 1 to 1000
time_values = []

for n in n_values:
    time_values.append(binary_search_time_simulation(n))

plt.figure(figsize=(10, 6))
plt.plot(n_values, time_values)
plt.xlabel("Input size n")
plt.ylabel("Time (log n)")
plt.title("O(log n) Logarithmic Time Complexity (Binary Search Simulation)")
plt.grid(True)
plt.show()
```

### Logarithmic Time Complexity O(log n)

An algorithm has logarithmic time complexity if its execution time grows proportionally to the logarithm of the input size. This often occurs when the algorithm repeatedly divides the problem size in half, such as in binary search.

**Binary Search Example:**
Binary search is an efficient algorithm for finding an item from a sorted list of items. It works by repeatedly dividing in half the portion of the list that could contain the item, until it narrows down the possible locations to just one. Each step effectively halves the search space, leading to logarithmic time complexity.

```python
import math
import matplotlib.pyplot as plt

# Prob_4: Write a function that runs in logarithmic time (Binary Search)
def binary_search_time_simulation(n):

    if n == 0:  # Handle edge case for log(0) which is undefined
        return 0
    return math.log2(n)

n_values = list(range(1, 1001)) # Input sizes from 1 to 1000
time_values = []

for n in n_values:
    time_values.append(binary_search_time_simulation(n))

plt.figure(figsize=(10, 6))
plt.plot(n_values, time_values)
plt.xlabel("Input size n")
plt.ylabel("Time (log n)")
plt.title("O(log n) Logarithmic Time Complexity (Binary Search Simulation)")
plt.grid(True)
plt.show()
```

#*Log-Linear Time Complexity O(n log n) : log-Â­linear running time*

```python
import math
import matplotlib.pyplot as plt

# Prob_5: Write a function that runs in log-linear time (Merge Sort simulation)
def log_linear_time_simulation(n):

    if n <= 1:
        return n
    return n * math.log2(n)

n_values = list(range(1, 1001))
time_values = []

for n in n_values:
    time_values.append(log_linear_time_simulation(n))

plt.figure(figsize=(10, 6))
plt.plot(n_values, time_values)
plt.xlabel("Input size n")
plt.ylabel("Time (n log n)")
plt.title("O(n log n) Log-Linear Time Complexity (Merge Sort Simulation)")
plt.grid(True)
plt.show()
```

### Polynomial Time Complexity O(n^c) : polynomial running time (c is a constant)

* When the computation time increases as function of N raised to some power, N being the size of input. Such a code has Polynomial Time complexity.




```python
import matplotlib.pyplot as plt

# Prob_6: Write a function that runs in polynomial time (O(n^c), here c=2 for O(n^2))
def polynomial_time_simulation(n):
    return n**2

n_values = list(range(1, 101)) # Input sizes from 1 to 100
time_values = []

for n in n_values:
    time_values.append(polynomial_time_simulation(n))

plt.figure(figsize=(10, 6))
plt.plot(n_values, time_values)
plt.xlabel("Input size n")
plt.ylabel("Time (n^2)")
plt.title("O(n^2) Polynomial Time Complexity Simulation")
plt.grid(True)
plt.show()
```

#Exponential Time Complexity O(c^n) : exponential running time (c is a constant being raised to a power based on size of input)

* When the computation time of a code increases as function of X^N, N being the size of input. Such a code has Polynomial Time Complexity.




```python
import math
import matplotlib.pyplot as plt

# Prob_7: Write a function that runs in exponential time (O(c^n), here c is approximately 1.618 for Fibonacci)
def exponential_time_simulation(n):

    return 2**n

n_values = list(range(1, 21))
time_values = []

for n in n_values:
    time_values.append(exponential_time_simulation(n))

plt.figure(figsize=(10, 6))
plt.plot(n_values, time_values)
plt.xlabel("Input size n")
plt.ylabel("Time (2^n)")
plt.title("O(2^n) Exponential Time Complexity Simulation")
plt.grid(True)
plt.show()
```

### Merge Sort Time Complexity (O(n log n))

Merge Sort is a stable, comparison-based sorting algorithm that uses the divide-and-conquer paradigm. It works by recursively dividing an unsorted list into `n` sublists, each containing one element (a list of one element is considered sorted). Then, it repeatedly merges sublists to produce new sorted sublists until there is only one sorted list remaining.

**Time Complexity Explanation:**
*   **Divide Step:** The division of the array into two halves can be represented by `log n` levels of recursion.
*   **Conquer/Merge Step:** At each level of recursion, the merging process involves comparing and combining elements, which takes linear time `O(n)`.
*   **Overall:** Since there are `log n` levels and each level involves `O(n)` work (merging), the total time complexity is `O(n log n)` in all cases (best, average, and worst).
